### Yêu cầu 2: 
- Sinh ngẫu nhiên một ma trận kích thước $10 \times 10$ để làm ma trận liên thuộc của một graph có hướng $10$ đỉnh.
- Tạo graph có hướng $H$ từ ma trận trên.
- Thực hiện vẽ graph $H$ (có mũi tên) theo đoạn code bên dưới:

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Sinh ngẫu nhiên ma trận kề
random_adjacency_matrix = np.random.randint(2, size=(10, 10))

# In ma trận kề
print("Ma trận kề ngẫu nhiên:")
print(random_adjacency_matrix)

# Tạo đồ thị có hướng từ ma trận kề
H = nx.DiGraph(random_adjacency_matrix)

# Vẽ đồ thị có hướng
plt.figure(figsize=(8, 8))
pos = nx.spring_layout(H)  # Lựa chọn layout cho đồ thị
nx.draw(H, pos, with_labels=True, arrows=True)
plt.title('Đồ thị có hướng với mũi tên')
plt.show()
```

```python
options = {
    'node_color': 'blue',
    'node_size': 100,
    'width': 3,
    'arrowstyle': '-|>',
    'arrowsize': 12,
}
nx.draw_networkx(H, arrows=True, with_labels=True, **options)
```

- Tham khảo link sau đây để chuyển graph trên sang dạng Json, thử đọc lại graph từ file Json lưu lại được:: 

https://networkx.org/documentation/stable/reference/readwrite/json_graph.html

- Tham khảo link sau đây để chuyển graph trên sang dạng GML, thử đọc lại graph từ file GML lưu lại được:

https://networkx.org/documentation/stable/reference/readwrite/gml.html

```python
import json

# Lưu đồ thị vào file JSON
json_data = nx.node_link_data(H)
with open('graph.json', 'w') as json_file:
    json.dump(json_data, json_file, indent=2)

# Lưu đồ thị vào file GML
nx.write_gml(H, 'graph.gml')

# Đọc lại đồ thị từ file JSON
with open('graph.json', 'r') as json_file:
    loaded_data = json.load(json_file)
    G_from_json = nx.node_link_graph(loaded_data)

# Đọc lại đồ thị từ file GML
G_from_gml = nx.read_gml('graph.gml')

# Kiểm tra xem đồ thị đã được đọc lại đúng chưa
print("Đồ thị từ file JSON:")
print(G_from_json.edges())

print("Đồ thị từ file GML:")
print(G_from_gml.edges())
```

- Tham khảo ý tưởng sau đây để in graph với đỉnh có màu, trong đó đỉnh bậc chẵn thì màu xanh, bậc lẻ thì màu đỏ.

```python
# Đọc file data.txt và tạo đồ thị
G = nx.read_edgelist("data.txt", nodetype=int)

# Sử dụng thuật toán nào đó để tính toán thông tin, ví dụ như giả sử bạn có một danh sách S như sau:

# Tính bậc của từng nút trong đồ thị
S = dict(G.degree())
# Tạo danh sách màu sắc dựa trên giá trị của S
colors = []
for v in G.nodes:
    if S[v] % 2 == 1:
        colors.append('green')
    else:
        colors.append('red')

# Vẽ đồ thị với màu sắc
nx.draw(G, with_labels=True, node_color=colors)
plt.show()
```

### Yêu cầu 3: 
- Viết hàm sinh ngẫu nhiên một graph với $n$ đỉnh và desity là $p$. Từ đó so sánh với hàm erdos_renyi_graph và gnm_random_graph có sẵn của thư viện.
- Vẽ tất cả các graph sau đây bằng cách tạo và vẽ graph ứng với thư viện có sẵn: cycle_graph, path_graph, star_graph, complete_graph, wheel_graph, full_rary_tree, grid_2d_graph, ego_graph, windmill_graph.

```python
import random
# Yêu cầu 3
def generate_random_graph(n, p):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)

    for i in nodes:
        for j in nodes:
            if i < j and random.uniform(0, 1) < p:
                G.add_edge(i, j)

    return G

# So sánh với erdos_renyi_graph và gnm_random_graph
n = 10  # Số đỉnh
p = 0.3  # Density

random_graph = generate_random_graph(n, p)
erdos_renyi_graph = nx.erdos_renyi_graph(n, p)
gnm_random_graph = nx.gnm_random_graph(n, n*(n-1)//2)

# Vẽ đồ thị
plt.figure(figsize=(15, 5))

plt.subplot(131)
nx.draw(random_graph, with_labels=True, font_weight='bold', node_color='skyblue', node_size=800)
plt.title('Random Graph')

plt.subplot(132)
nx.draw(erdos_renyi_graph, with_labels=True, font_weight='bold', node_color='lightcoral', node_size=800)
plt.title('Erdos Renyi Graph')

plt.subplot(133)
nx.draw(gnm_random_graph, with_labels=True, font_weight='bold', node_color='lightgreen', node_size=800)
plt.title('GNM Random Graph')

plt.show()

# Vẽ các đồ thị khác
graphs = [
    nx.cycle_graph(5),
    nx.path_graph(5),
    nx.star_graph(5),
    nx.complete_graph(5),
    nx.wheel_graph(5),
    nx.full_rary_tree(2, 5),
    nx.grid_2d_graph(3, 3),
    nx.ego_graph(random_graph, 0),
    nx.windmill_graph(3, 4)
]

plt.figure(figsize=(15, 10))
for i, G in enumerate(graphs, 1):
    plt.subplot(3, 3, i)
    nx.draw(G, with_labels=True, font_weight='bold', node_color='lightgray', node_size=800)
    plt.title(type(G).__name__)

plt.show()
```

### Yêu cầu 4:
- Cho danh sách tên người sau đây: Anh, Đức, Dũng, Hòa, Huy, Kiệt, Minh, Nam, Nghĩa, Tân. Hãy sử dụng hàm đã viết ở câu trên để sinh ra một graph với $p=0.7$, trong đó các đỉnh ứng với tên $10$ người.
- Đọc nội dung lý thuyết sau đây và trả lời các câu hỏi: tính số đường đi từ Nam đến Huy với độ dài đường đi là $5$, tính số bộ ba người đôi một quen nhau, kiểm tra tính liên thông của graph này.

```python
#CODE HERE
# Yêu cầu 4
names = ["Anh", "Đức", "Dũng", "Hòa", "Huy", "Kiệt", "Minh", "Nam", "Nghĩa", "Tân"]
random_graph_names = generate_random_graph(len(names), 0.7)
mapping = dict(zip(range(len(names)), names))
random_graph_names = nx.relabel_nodes(random_graph_names, mapping)

# Vẽ đồ thị với tên người
plt.figure(figsize=(8, 8))
nx.draw(random_graph_names, with_labels=True, font_weight='bold', node_color='skyblue', node_size=800)
plt.title('Random Graph with Names')
plt.show()

# Tính số đường đi từ Nam đến Huy với độ dài là 5
num_paths = nx.all_simple_paths(random_graph_names, source="Nam", target="Huy", cutoff=5)
print("Số đường đi từ Nam đến Huy có độ dài là 5:", len(list(num_paths)))

# Tính số bộ ba người đôi một quen nhau
triads = nx.triadic_census(random_graph_names.to_directed())
print("Số bộ ba người đôi một quen nhau:", triads["003"] / 3)



# Kiểm tra tính liên thông của đồ thị
is_connected = nx.is_connected(random_graph_names)
print("Tính liên thông của đồ thị:", is_connected)
```
